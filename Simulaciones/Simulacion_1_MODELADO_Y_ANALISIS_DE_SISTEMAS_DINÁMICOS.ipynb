## SISTEMA DE TANQUES

```python
#definición del sistema de tanques
num=[0.3188]
den=[1., 1.143, 0.3188]
ftla= tf(num, den)
print('Funcion de transferencia en LA de la planta: ')
print(ftla)
ftlc=feedback(ftla,1)
print("Funcion de transferencia en lazo cerrado de la planta: ")
print(ftlc)
```

```python
# Gráfico de polos y ceros del sistema en lazo cerrado.
plt.figure(1, figsize=[7, 4])
pzmap(ftlc)
plt.xlabel('Real')
plt.ylabel('Imaginario')
plt.title('Plano complejo. Polos del sistema en LC')
plt.show(block=False)
```

```python
# Respuesta escalón del sistema en LA y LC
plt.figure(2, figsize=[6, 4])

# Respuesta escalón
yout, T = step(ftla, T=25)
yout1, T1 = step(ftlc, T=25)

# Gráfica
plt.plot(T, yout, color='red', linewidth=2, label='Sistema en LA')
plt.plot(T1, yout1, color='darkgreen', linewidth=2, label='Sistema en LC')

plt.xlabel('Tiempo (s)')
plt.ylabel('$y(t)$')
plt.title('Respuesta escalón: Sistema en LA vs LC')
plt.grid(True)
plt.legend()
plt.tight_layout()
```

```python
# Respuesta impulso del sistema en LA y LC
plt.figure(3, figsize=[6, 4])

# Respuesta impulso
yout, T = impulse(ftla, T=25)
yout1, T1 = impulse(ftlc, T=25)

# Grafica
plt.plot(T, yout, color='red', linewidth=2, label='Sistema en LA')
plt.plot(T1, yout1, color='darkgreen', linewidth=2, label='Sistema en LC')

plt.xlabel('Tiempo (s)')
plt.ylabel('$y(t)$')
plt.title('Respuesta impulso: Sistema en LA vs LC')

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
